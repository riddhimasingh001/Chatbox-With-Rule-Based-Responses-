```python
import random
import re
from datetime import datetime

memory = {
    "name": None,
    "mood": None
}

def get_weather():
    return "I can't fetch real-time weather yet, but I hope it's nice where you are! ‚òÄÔ∏è"

def get_math_answer(expression):
    try:
        result = eval(expression)
        return f"That equals {result}"
    except:
        return "Hmm, I couldn't calculate that. Try a simpler math expression."

def chatbot():
    print("Chatbot: Hey there! I'm SmartBot ü§ñ. Ready to chat, help, or make you smile üòä. Type 'bye' to exit.")

    jokes = [
        "Why don‚Äôt scientists trust atoms? Because they make up everything!",
        "Why did the math book look sad? Because it had too many problems.",
        "I'm reading a book on anti-gravity. It's impossible to put down!"
    ]

    facts = [
        "Sloths can hold their breath longer than dolphins.",
        "Octopuses have three hearts and blue blood.",
        "Bananas are berries, but strawberries aren‚Äôt."
    ]

    advice_list = [
        "Stay positive and keep moving forward!",
        "Take regular breaks while studying or working.",
        "Keep a gratitude journal‚Äîit really helps with mental clarity."
    ]

    songs = [
        "'Blinding Lights' by The Weeknd",
        "'Levitating' by Dua Lipa",
        "'Perfect' by Ed Sheeran",
        "'Shape of You' by Ed Sheeran"
    ]

    movies = [
        "Inception (2010) - Sci-Fi",
        "The Shawshank Redemption (1994) - Drama",
        "Interstellar (2014) - Sci-Fi",
        "Inside Out (2015) - Animation"
    ]

    fallback_responses = [
        "That's an interesting point. Could you explain a bit more?",
        "I'd love to know more about what you're thinking!",
        "Can you tell me a little more about that?",
        "That sounds important‚Äîgo on, I'm listening."
    ]

    while True:
        user_input = input("You: ").lower().strip()

        if user_input in ["exit", "bye", "goodbye", "quit"]:
            name = memory['name'] or 'friend'
            print(f"Chatbot: Goodbye {name}! Take care and talk soon üíõ")
            break

        if "my name is" in user_input:
            name_match = re.search(r"my name is (\w+)", user_input)
            if name_match:
                memory['name'] = name_match.group(1).capitalize()
                print(f"Chatbot: Great to meet you, {memory['name']}! üòä")
            continue

        if "i feel" in user_input:
            mood_match = re.search(r"i feel (\w+)", user_input)
            if mood_match:
                memory['mood'] = mood_match.group(1)
                print(f"Chatbot: Got it ‚Äî you're feeling {memory['mood']}. Want to talk about it?")
            continue

        if re.search(r"\b(hi|hello|hey|yo|sup)\b", user_input):
            greeting = f"Hey {memory['name']}!" if memory['name'] else "Hey there!"
            print(f"Chatbot: {greeting} How can I brighten your day today?")
            continue

        if "how are you" in user_input:
            print("Chatbot: I‚Äôm all good! What about you? üòä")
            continue

        if re.search(r"\b(thanks|thank you|thx|ty|ok thanks|okay thanks|oh thanks|ohh thanks)\b", user_input):
            print("Chatbot: You're very welcome! üíõ")
            continue

        if re.search(r"\b(nice|cool|great|awesome|sweet|amazing|lovely)\b", user_input):
            print("Chatbot: Glad you think so! üòÑ")
            continue

        if re.search(r"\b(yes|yeah|yup|sure|of course|yess)\b", user_input):
            print("Chatbot: Awesome! Let‚Äôs go on ‚ú®")
            continue

        if re.search(r"\b(no|nope|nah|not really|never mind)\b", user_input):
            print("Chatbot: No worries, we can switch it up! üòä")
            continue

        if "weather" in user_input:
            print("Chatbot:", get_weather())
            continue

        if re.search(r"\b(date|day|today)\b", user_input):
            print("Chatbot: Today is", datetime.today().strftime("%A, %B %d, %Y"))
            continue

        if "time" in user_input:
            print("Chatbot: It‚Äôs", datetime.now().strftime("%I:%M %p"))
            continue

        if re.search(r"\b(joke|funny|laugh)\b", user_input):
            print("Chatbot:", random.choice(jokes))
            continue

        if re.search(r"\b(fact|did you know|interesting)\b", user_input):
            print("Chatbot:", random.choice(facts))
            continue

        if re.search(r"\b(advice|tip|suggestion)\b", user_input):
            print("Chatbot:", random.choice(advice_list))
            continue

        if re.search(r"\b(song|music|listen|track)\b", user_input):
            print("Chatbot: How about listening to", random.choice(songs))
            continue

        if re.search(r"\b(movie|film|watch|cinema)\b", user_input):
            print("Chatbot: You might like:", random.choice(movies))
            continue

        if re.search(r"\d+[\+\-\*/]\d+", user_input):
            print("Chatbot:", get_math_answer(user_input))
            continue

        print("Chatbot:", random.choice(fallback_responses))

chatbot()


```
